## Lecture 11




Ineq Constraint:

active: $a_i^Tx=b_i$ @ solution\

inactive $a_i^Tx > b_i$ @ solution

The active set is the set of active constraint.

Let $x^*$ be a solution of 
$$
\min f(x), s.t. Ax \ge b <=> \min f(x), s.t. \tilde{A}x=\tilde{b}
$$
where $\hat{A}$, $\hat{b}$ are the rows of $A$, $b$ for active constraints.

This gives us necessary condition for a local minima.

$$
\begin{split}
\hat{z}^Tg(x^*)=0, \hat{z} =  null(\hat{A}) \\
\hat{A}^T\hat{\lambda}=g(x^*) \\
Ax^* \ge b \\
\hat{z}^TH(x^*)\hat{z} \ge 0 \\
\hat{\lambda} \ge 0
\end{split}\
$$

The term **descent direction** is used for any vector $\rho, s.t. \rho^Tg(x)<0$.

Line Search: $x_{k+1}=x_k+\alpha\rho_k$
Why descent direction?

$$
f(x_k+\alpha_k\rho_k)=f(x_k)+\alpha_k\rho_k^Tg_k+O(\alpha^2)
$$

so if $\rho, s.t. \rho^Tg(x)<0$, then for some tiny $\alpha_k$, $f(x_{k+1}) < f(x_k)$.
