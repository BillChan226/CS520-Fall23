### Steepest Gradient Descent



Let's solve 

$$
\min_{\alpha_k} f(x_k+\alpha_kg_k)
$$

$$
f(x_{k+1}) \le f(x_k)(1-c)
$$
for some $0<c<1$ => $f(x_k)\to 0$, as $k \to \infty$ at least at a geometric/linear rate.

1. at $x_k$, we have $g_k=g(x_k)=Qx_k$
   
   (if $Q!=Q^T$, then $g(x)=(\frac{1}{2}Q+\frac{1}{2}Q^T)x$).
2. $\min_{\alpha_k} f(x_k-\alpha_kg_k)=\frac{1}{2}(x_k-\alpha g_k)^TQ(x_k-\alpha g_k)$



$$
f(x_{k+1})=f(x_k)-\frac{1}{2}\frac{||g(k)||^4}{g_k^TQg_k}
$$

since we have $f(x_k)=\frac{1}{2}g_k^TQ^{-1}g_k$
then 
$$
f(x_{k+1})=f(k_k)\times\{1-\frac{||g(k)||^4}{g_k^TQg_kg_k^TQ^{-1}g_k}\}
$$
